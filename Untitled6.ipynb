```python

df_raw = pd.read_csv("sales_raw - sales_raw.csv")

print("First 5 rows:")
df_raw.head()
```

The dataset contains issues such as incorrect date formats, non-numeric prices, missing values, zero or negative quantities/prices, and duplicate sale records.

```python
raw_record_count = len(df_raw)
print("Total number of raw records:", raw_record_count)
```

This step converts transaction dates to datetime format and ensures quantity and unit price fields are numeric by handling parsing errors safely.

```python
print("Identified Data Quality Issues:")
print("1. transaction_date is not in datetime format")
print("2. unit_price contains non-numeric symbols ($)")
print("3. Missing values in qty or unit_price")
print("4. Zero or negative qty and unit_price")
print("5. Duplicate sale_id values")
```

Records with missing, zero, or negative quantity or price values are removed because they cannot generate valid revenue and distort reporting results.

```python
df = df_raw.copy()

df["transaction_date"] = pd.to_datetime(df["transaction_date"], errors="coerce")

df["qty"] = pd.to_numeric(df["qty"], errors="coerce")

df["unit_price"] = df["unit_price"].astype(str).str.replace("$", "", regex=False)
df["unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")

df.head()
```

Duplicate sales are removed using the sale_id field to prevent double counting of orders and revenue in reports.

```python

df = df.dropna(subset=["transaction_date", "qty", "unit_price"])


df = df[(df["qty"] > 0) & (df["unit_price"] > 0)]

print("Records after removing invalid values:", len(df))
```

A clean fact table is created by calculating revenue for each sale, ensuring one valid transaction per row with correct data types.

```python

before_dedup = len(df)

df = df.drop_duplicates(subset="sale_id", keep="first")

after_dedup = len(df)

print("Records before deduplication:", before_dedup)
print("Records after deduplication:", after_dedup)

```

```python
df["revenue"] = df["qty"] * df["unit_price"]

print("FACT TABLE STRUCTURE:")
df.info()
```

A clean fact table is created by calculating revenue for each sale, ensuring one valid transaction per row with correct data types.

```python
clean_record_count = len(df)
retention_percentage = (clean_record_count / raw_record_count) * 100

print("Raw Records:", raw_record_count)
print("Clean Records:", clean_record_count)
print("Percentage Records Retained:", round(retention_percentage, 2), "%")
```
